```python
from safetensors.torch import load_file, save_file

tensors = load_file("../checkpoints/smoltts_1227_49000/model.safetensors")
tensors.keys()
```

```python
# Modify the keys by replacing '_orig_mod' with 'model'
updated_tensors = {key.replace("_orig_mod.", ""): value for key, value in tensors.items()}
updated_tensors.keys()
```

```python

# Save the updated tensors to a new file
updated_file_path = "../checkpoints/smoltts_1227_49000/model.safetensors"
save_file(updated_tensors, updated_file_path)

print(f"Updated safetensors saved to {updated_file_path}")
```
